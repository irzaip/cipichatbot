```python
import requests
from bs4 import BeautifulSoup
import time
```

```python
urld="https://www.goodreads.com/quotes?page=%s"
```

```python
import re

def cleanhtml(raw_html):
  cleanr = re.compile('<.*?>')
  cleantext = re.sub(cleanr, '', raw_html)
  return cleantext
```

```python
def dslash_remove(raw_html):
  cleanr = re.compile('\/\/[.\s\w\W]+\/\/]]>')
  cleantext = re.sub(cleanr, '', raw_html)
  return cleantext
```

```python
def scrape_all(totpage):
    for i in range(totpage):
        ress = extract(scrp(i))
        print("Extract page:",i)
        time.sleep(2)
```

```python
def scrp(ppage):
    pg = requests.get(urld % ppage)
    result = BeautifulSoup(pg.text,'lxml')
    return result
```

```python
def extract(pagequote):
    global result
    _reslt = pagequote.find_all('div',{'class':'quoteText'})
    for rslt in _reslt:
        result.append(dslash_remove(rslt.get_text()))
    return result
```

```python
result=[]
```

```python
scrape_all(100)
```

```python
with open('quotes.txt','w') as f:
    for res in result:
        try:
            f.writelines(res)
        except:
            print(res)
            pass
```
