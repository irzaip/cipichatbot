```python
def translate(sentence):
    clearsrc()
    srcbox = browser.find_element_by_id("source")
    time.sleep(0.5)
    srcbox.send_keys(sentence)
    time.sleep(1)
    rsltbox = browser.find_element_by_class_name("tlid-copy-target")
    time.sleep(0.5)
    reslt = rsltbox.text
    time.sleep(0.5)
    return reslt
    
```

```python
def clearsrc():
    srcbox = browser.find_element_by_id("source")
    srcbox.send_keys(Keys.CONTROL+"a")
    srcbox.send_keys(Keys.DELETE)
    time.sleep(0.4)
```

# RUTIN UNTUK SPLIT FILE SEJUMLAH KATA < 5000 untuk google translate

```python
import nltk

def splitfile(filename,wordlength=5000, maxline=300):
    """memotong file menjadi beberapa file kecil dengan jumlah baris -> maxline"""
    fs=open(filename,"r",encoding="utf8").readlines()
    res = []  #hasil dimasukan ke sini
    lf = []
    wc = 0
    rl = 0
    fn = 1
    for i in range(len(fs)):
        rl += 1
        wc = wc + len(nltk.word_tokenize(fs[i]))
        if wc>wordlength:
            print("line exceed 5000")
        res.append(fs[i])
        if rl == maxline:
            fnm = filename+str(fn)+".txt"
            with open(fnm,"w") as f:
                for i in range(len(res)):
                    f.writelines(res[i])
            f.close()
            res=[]
            print("Write file:",fnm)
            fn += 1
            rl=0
            wc=0
```

```python
def translate_file(filename):
    browser.get("https://translate.google.com/?tr=f&sl=en&tl=id")


    import os
    filebtn = browser.find_element_by_id("file")
    #filebtn.click()

    filebtn.send_keys(os.getcwd()+"/"+filename)
    submitbtn = browser.find_element_by_id("gt-submit")
    submitbtn.click()

    from bs4 import BeautifulSoup
    sp = BeautifulSoup(browser.page_source,"html.parser")

    af = sp.find("pre").get_text()

    tpt = af.split("\n")

    #pop untuk ngurangin yang terakhir
    tpt.pop()

    fnn = "tj-"+filename
    with open(fnn,'w',encoding="utf8") as f:
        for i in tpt:
            f.writelines(i+"\n")
    print("Done writing:",fnn)
    f.close()
```

```python
for i in range(0,69):
    translate_file(frd[i])
```
