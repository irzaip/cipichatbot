```python
import re
import time
from bs4 import BeautifulSoup
import requests
```

```python
urld = "http://kateglo.com/?&phrase=&mod=proverb&p=%s"
```

```python
def scrape_all(totpage):
    for i in range(totpage):
        ress = extract_text(scrp(i))
        print("Extract page:",i)
        time.sleep(2)
```

```python
def dslash_remove(raw_html):
  cleanr = re.compile('\/\/[.\s\w\W]+\/\/]]>')
  cleantext = re.sub(cleanr, '', raw_html)
  return cleantext
```

```python
def scrp(ppage):
    pg = requests.get(urld % ppage)
    result = BeautifulSoup(pg.text,'lxml')
    return result
```

```python
def extract_text(pagequote):
    global result
    _reslt = pagequote.find_all('dt')
    _resl2 = pagequote.find_all('dd')
    for rslt,rslt2 in zip(_reslt,_resl2):
        result.append(dslash_remove(rslt.get_text()))
        result.append(dslash_remove(rslt2.get_text()))
    return result
```
