```python
import re
import glob
import nltk
```

```python
srts = glob.glob('../data/subscene/*.srt')
```

```python
def writetxt(fname,data):
    with open(fname, "w") as fn:
        for i in data:
            fn.writelines(i+'\n')
    print("Done writing:", fname)
```

```python
def clean(data):
    result = re.sub(r'\W+_', '', data)
    return result
```

```python
def cleantag(data):
  cleanr = re.compile('<.*?>')
  cleantext = re.sub(cleanr, '', data)
  return cleantext
```

```python
def remove_non_ascii_2(text):
    return re.sub(r'[^\x00-\x7F]','', text)
```

```python
def process_srt(fname):
    f = open(fname,"r")
    dt = f.readlines()
    f.close()

    #ambil yang dialog saja
    result=[]
    for i in range(len(dt)):
        if dt[i]=='\n': 
            txt = dt[i-1]
            txt = txt.replace("\n","")
            txt = remove_non_ascii_2(txt)
            result.append(clean(cleantag(txt)))
    writetxt(fname+".txt",result)
```

```python
for i in srts:
    try:
        process_srt(i)
    except Exception as e:
        print(e)
```
