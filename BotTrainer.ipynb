```python
def swv(sentence, rempunct=False, flat=False):
    """Sentence to word2vec"""
    if not flat:
        sntoken = nltk.sent_tokenize(sentence)
    else:
        sntoken = [sentence]
    for i in range(len(sntoken)):
        tokens = nltk.word_tokenize(sntoken[i])
        if rempunct==True:
            type(tokens)
            text = nltk.Text(tokens)
            type(text)  
            sntoken[i] = [w.lower() for w in text if w.isalpha()]
        else: sntoken[i] = tokens
    if len(sntoken)==1: 
        sntoken = sntoken[0]
        remove_unchar(sntoken)
    for i in sntoken:
        remove_unchar(i)
    if sntoken==[]: sntoken=['yang']
    print(sntoken)
    return sntoken
```

```python
def swv_ar(sentence, maxword=50, vecsize=200):
    """Sentence word2vec to array matrices for processing"""
    senarray=[]
    if type(sentence[0])==list:
        for i in range(len(sentence)):
            for k in range(len(sentence[i])):
                #print(tex2vec[sentence[i][k]])
                try:
                    senarray.append(np.array(tex2vec[sentence[i][k]]))
                except:
                    pass
    else:
        for i in range(len(sentence)):
            try:
                senarray.append(np.array(tex2vec[sentence[i]]))
            except:
                pass
    
    senarray.reverse()
    for i in range(maxword-len(senarray)):
        senarray.append(np.zeros(vecsize))
    senarray.reverse()
    return np.array(senarray)
```
