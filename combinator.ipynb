# Combine multiple files to one single file

```python
import glob
import os
```

```python
filedir="../data/terjemah/"
filename='tj-train.dec'
```

```python
with open("train.dec-id.txt","w",encoding="utf8") as f:
    for i in range(1,541):
        fproc = filedir+filename+str(i)+".txt"
        print(fproc)
        f.writelines(open(fproc,'r',encoding="utf8").readlines())
```
