```python
import selenium
import time
from bs4 import BeautifulSoup
import re
import time
```

```python
from selenium import webdriver 
from selenium.webdriver.common.by import By 
from selenium.webdriver.support.ui import WebDriverWait 
from selenium.webdriver.support import expected_conditions as EC 
from selenium.common.exceptions import TimeoutException
```

```python
option = webdriver.ChromeOptions()
#option.add_argument("--incognito")
browser = webdriver.Chrome(chrome_options=option)
```

```python
browser.get("http://subscene.com/browse")
```

# PARSE DARI SINI

```python
getlinks=[]
def getlist():
    global getlinks
    pg = BeautifulSoup(browser.page_source,"html.parser")

    lss = pg.find_all("a",href=True)

    getlinks=[]
    for i in range(len(lss)):
        if lss[i].get_text().find("Indonesia")>=0: getlinks.append(lss[i])

    #buang yg terakhir
    getlinks.pop()
```

# LOOP SCRAPING

```python
def leech():
    for i in range(len(getlinks)):
        url=getlinks[i]['href']
        url="http://subscene.com"+url
        print(url)
        browser.get(url)
        
        try:
            dl = browser.find_element_by_link_text("Download Indonesian Subtitle")
            dl.click()
        except:
            pass
        browser.back()
    print("DONE-Leeching")
```

#### Tentukan halaman di range dibawah ini.

```python
for i in range(45,55):
    url='https://subscene.com/browse/latest/all/'+str(i)
    browser.get(url)
    getlist()
    leech()
    print("Done this page ----")
```
